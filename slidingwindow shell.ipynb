```python
for j in range(3,9): 
    d = []
    e = []
    for i in range(len(df)):
        
        value = (df.iloc[i,j]+df.iloc[i,j+1]+df.iloc[i,j+2])/3
        fore1 = (0.1*df.iloc[i,j+2]) + (0.9*value)
        fore2  = (0.1*df.iloc[i,j+3]) + (0.9*fore1)
        d.append(fore1)
        e.append(fore2)
        
    df['fore'+ str(j)] = d
    df['fore'+str(j)+str(j)] = e
```
